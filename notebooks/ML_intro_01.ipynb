```python
iris.groupby('species').mean()
```

```python
iris.plot(kind='scatter', x='sepal_length', y='sepal_width')
plt.title('Iris Sepal Length vs Width')
```

```python
iris.plot(kind='hist', y='petal_length', bins=20)
plt.title('Iris Petal Length Distribution')
```
